```python
from mnist import MNIST
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Spliting data into two sets: training and testing:

```python
mndata = MNIST('.')
images, labels = mndata.load_training()
images = np.array(images)
labels = np.array(labels)
```

## 2. Display images with their labels:

```python
num_images = 10
fig, axes = plt.subplots(2, 5, figsize=(12, 6))
axes = axes.flatten()

for i in range(num_images):
    ax = axes[i]
    img = images[i].reshape(28, 28)
    ax.imshow(img, cmap='gray')
    ax.set_title(f'Label: {labels[i]}')
    ax.axis('off')

plt.tight_layout()
plt.show()
```

## 3. Creating training model:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
model = LogisticRegression()

X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)

model.fit(X_train, y_train)
```

## 4. Evaluating model's performance:

```python
model.score(X_test, y_test)
```

## 5. Predicting values:

```python
single_image = images[1].reshape(1, -1)

img = images[1].reshape(28, 28)
plt.imshow(img, cmap='gray')
plt.title(f'Label: {labels[1]}')
plt.axis('off')

plt.tight_layout()
plt.show()

prediction = model.predict(single_image)
print(f"The predicted label of the giving image is: {prediction}")
```

## 6. Displaying model's mistakes:

```python
y_pred = model.predict(X_test)

misclassified = (y_test != y_pred)
X_wrong = X_test[misclassified]
y_wrong_true = y_test[misclassified]
y_wrong_pred = y_pred[misclassified]

def display_digit(image):
    plt.imshow(image.reshape(28, 28), cmap=plt.cm.binary)
    plt.axis('off')
    plt.show()

for i in range(4):
    display_digit(X_wrong[i].reshape(28, 28))  # Reshape for display
    print(f"Predicted: {y_wrong_pred[i]}, Actual: {y_wrong_true[i]}")
```
