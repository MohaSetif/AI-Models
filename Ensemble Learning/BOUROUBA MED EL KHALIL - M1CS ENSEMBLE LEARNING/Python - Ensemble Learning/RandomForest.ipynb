```python
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier
from sklearn.datasets import load_digits
from sklearn.tree import DecisionTreeClassifier
import numpy as np
from sklearn.model_selection import train_test_split

digits = load_digits()
X, y = digits.data, digits.target
clf = DecisionTreeClassifier()
clf.fit(X, y)
accuracy = clf.score(X,y)
print(accuracy)
```

```python
bagging = BaggingClassifier(DecisionTreeClassifier(), n_estimators=200)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.90)
bagging.fit(X_train, y_train)
bagging_accuracy = bagging.score(X_test, y_test)

rf = RandomForestClassifier(n_estimators=200)
rf.fit(X_train, y_train)
rf_accuracy = rf.score(X_test, y_test)

print(f"BaggingClassifier Accuracy : {bagging_accuracy}")
print(f"RandomForestClassifier Accuracy : {rf_accuracy}")
```

```python
accuracies = []
for i in range(100):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9)
    rf = RandomForestClassifier(n_estimators=200)
    rf.fit(X_train, y_train)
    accuracy = rf.score(X_test, y_test)
    accuracies.append(accuracy)

rf_variance = np.var(accuracies)
print(f"RandomForestClassifier's Accuracy Variance : {rf_variance}")
```

```python
import matplotlib.pyplot as plt

n_estimators_values = range(1, 201)
accuracies1 = []

for n_estimators in n_estimators_values:
    rf = RandomForestClassifier(n_estimators=n_estimators)
    rf.fit(X_train, y_train)
    accuracy = rf.score(X_test, y_test)
    accuracies1.append(accuracy)

plt.plot(n_estimators_values, accuracies1)
plt.xlabel("Estimators' Number")
plt.ylabel('Accuracy')
plt.show()
```

```python
from sklearn.ensemble import ExtraTreesClassifier

extra_trees = ExtraTreesClassifier(n_estimators=200)
extra_trees.fit(X_train, y_train)
extra_trees_accuracy = extra_trees.score(X_test, y_test)

print(f"Accuracy de RandomForestClassifier : {rf_accuracy}")
print(f"Accuracy de ExtraTreesClassifier : {extra_trees_accuracy}")
```
