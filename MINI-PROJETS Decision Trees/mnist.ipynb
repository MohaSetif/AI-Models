```python
from mnist import MNIST
import numpy as np
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

mndata = MNIST('.')
images, labels = mndata.load_training()
images = np.array(images)
labels = np.array(labels)

print(f"Information about the images : {images.shape}")
print(f"Information about the labels : {labels.shape}")

x_train, x_test, y_train, y_test = train_test_split(images, labels, test_size=0.3)

model1 = DecisionTreeClassifier(criterion='entropy', max_depth=2)
model1.fit(x_train, y_train)
print(f"The first model's score: {model1.score(x_train, y_train)}")

model2 = DecisionTreeClassifier(criterion='gini', max_depth=2)
model2.fit(x_train, y_train)
print(f"The second model's score: {model2.score(x_train, y_train)}")

logistic_model = LogisticRegression()
logistic_model.fit(x_train, y_train)
print(f"The logistic model's score: {logistic_model.score(x_train, y_train)}")


# Visualiser l'arbre de décision avec critère d'entropie
fig, ax = plt.subplots(figsize=(20, 10))
tree.plot_tree(model1, max_depth=2, feature_names=range(784), filled=True, ax=ax)
ax.set_title("Decision tree with entropy")

# Visualiser l'arbre de décision avec critère de Gini
fig, ax = plt.subplots(figsize=(20, 10))
tree.plot_tree(model2, max_depth=2, feature_names=range(784), filled=True, ax=ax)
ax.set_title("Decision tree with Gini")
```

The decision tree models performed poorly on the MNIST handwritten digit classification task, achieving accuracies of only around 0.34 due to their inability to effectively handle the high-dimensional data. The visualizations revealed that the trees relied heavily on individual pixel features, failing to capture higher-level patterns. In contrast, logistic regression substantially outperformed the decision trees, attaining an accuracy of 0.937, highlighting its suitability for such high-dimensional problems. The decision tree visualizations and accuracy scores clearly demonstrated the limitations of this model for the MNIST dataset and the superiority of logistic regression for this particular task.
